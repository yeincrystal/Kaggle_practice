## 자동으로 Categorical 변수 처리하기 -> Label Encoder



```python
from sklearn.preprocessing import LabelEncoder
```

```python
le = LabelEncoder() #library는 ()필요, 여기선 옵션 안 넣어줌 
all_data2['Make'] = le.fit_transform(all_data2['Make'])  #le라는 library안에 있는 fit_transform이라는 함수를 사용해서 labeling한 것 
```

```python
le.classes_
```

```python
columns = all_data2.columns
c = columns[all_data2.dtypes == object]
```

```python
c
```

```python
for i in c:
    all_data2[i] = le.fit_transform(list(all_data2[i])) #NaN이라는 값이 리스트에 들어가면 NaN이라는 문자로 바뀜
```

```python
all_data2
```

```python
#숫자형 데이터의 결측치 처리
all_data2 = all_data2.fillna(-1)
#-1을 넣는 이유? 
#-> 결측치 자체의 느낌 자체를 학습할 수 있도록 하기 위해서 (평균을 넣으면 결측치라는 특성을 반영하지 못하게 됨)-> 실제 test 셋에서도 결측치를 잘 반영해서 예측하게 됨 
#-> 선형 모델을 사용할 때는(Not 트리모델) 숫자의 특성을 (관계성) 가져오기 때문에 -1, 999 이런 값을 넣으면 모델 자체에 왜곡 일어날 수 있음 
```

```python
train2 = all_data2[:len(train)]
test2 = all_data2[len(train):]
```

```python
#random Foreast는 시간 bar 못 봄 
```

```python
%%time
from sklearn.ensemble import RandomForestClassifier
rf = RandomForestClassifier(n_jobs=-1, random_state=42)
rf.fit(train2,train['IsBadBuy'])
result = rf.predict_proba(test2)
```
